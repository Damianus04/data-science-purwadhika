## SCATTER PLOT



```python
# Import Matplotlib & Seaborn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import dataset from seaborn
tips = sns.load_dataset('tips')
tips.head(5)
```

#### Matplotlib

```python
# figure & axis
fig = plt.figure(figsize=(8,6))
ax = plt.axes()

# data
total_bill = tips['total_bill']
tip = tips['tip']

ax.scatter(total_bill, tip)

# data label
plt.title('Scatter Plot: Total Bill vs Tip', size=15)
plt.xlabel('Total Bill')
plt.ylabel('Tip')

# show
plt.show()
```

#### Seaborn

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day')

plt.title('Scatter Plot: Total Bill vs Tip', size=15)

plt.show()
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip')

plt.title('Scatter Plot: Total Bill vs Tip', size=15)

plt.show()
```

#### Pandas

```python
tips.plot.scatter(x='total_bill', y='tip')

plt.title('Scatter Plot: Total Bill vs Tip', size=15)

plt.show()
```

## BARPLOT

```python
# Import Matplotlib & Seaborn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import dataset from seaborn
tips = sns.load_dataset('tips')
tips.head(5)
```

```python
tips_by_day = tips[['tip', 'day']].groupby(['day'], as_index=True).mean()
tips_by_day
```

#### Matplotlib - vertical

```python
# figure & axis
fig = plt.figure(figsize=(8,6))
ax = plt.axes()

# data
day = tips['day'].unique().tolist()
tip = tips_by_day['tip']

ax.bar(day, tip, color = ['g', 'b', 'r', 'y'])

# data label
plt.title('Barplot: Average Tips Everyday', size=15)
plt.xlabel('Days')
plt.ylabel('Tip')

# show
plt.show()
```

#### Matplotlib - horizontal

```python
# figure & axis
fig = plt.figure(figsize=(8,6))
ax = plt.axes()

# data
day = tips['day'].unique().tolist()
tip = tips_by_day['tip']

ax.barh(day, tip, color = ['g', 'b', 'r', 'y'])

# data label
plt.title('Barplot: Average Tips Everyday', size=15)
plt.xlabel('Days')
plt.ylabel('Tip')

# show
plt.show()
```

#### Seaborn

```python
sns.barplot(data=tips, x='day', y='tip', hue='sex')

plt.title('Barplot: Average Tips Everyday & Gender', size=15)

plt.legend(loc=0)

plt.show()
```

#### Seaborn - horizontal

```python
sns.barplot(data=tips, x='tip', y='day', hue='sex')

plt.title('Barplot: Average Tips Everyday & Gender', size=15)

plt.legend(loc=0)

plt.show()
```

#### Pandas - horizontal

```python
tips_by_day = tips[['tip', 'day']].groupby(['day'], as_index=False).mean()
tips_by_day
```

```python
tips_by_day.plot.bar(x='day', y='tip')
plt.title('Total Bill: Dinner vs Lunch', size=15)
plt.show()
```

#### Pandas - horizontal

```python
tips_by_day.plot.barh(x='day', y='tip')
plt.title('Total Bill: Dinner vs Lunch', size=15)
plt.show()
```
